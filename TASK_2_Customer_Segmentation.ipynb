```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# prompt: print values of dataset

df = pd.read_csv(list(uploaded.keys())[0])
df.head(200)
```

```python
df.drop(["CustomerID"], axis = 1, inplace=True)
```

```python
df.isnull().sum()
```

```python
features = df[["Age", "Annual Income (k$)"]]
features
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['Age'], df['Annual Income (k$)'])
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
```

```python
kmeans.fit(features)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['Age'], df['Annual Income (k$)'], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, color='red')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df['Annual Income (k$)'] = scaler.fit_transform(df[['Annual Income (k$)']])
```

```python
df['Age'] = scaler.fit_transform(df[['Age']])
```

```python
features = df[["Age", "Annual Income (k$)"]]
features
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
```

```python
kmeans.fit(features)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['Age'], df['Annual Income (k$)'])
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['Age'], df['Annual Income (k$)'], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, color='red')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('KMeans Clustering')
plt.show()
```
